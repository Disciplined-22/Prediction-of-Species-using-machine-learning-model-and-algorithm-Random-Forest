```python
import seaborn as sns

iris = sns.load_dataset('iris')
print(iris)
```

```python
_ = sns.scatterplot(x='sepal_width', y='petal_length', hue='species', data=iris)
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]


X
```

```python
y = iris.species
y
```

```python
!pip install scikit-learn

!pip install numpy

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# iris_dt = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot individual trees from the Random Forest
n_trees_to_plot = 3  # You can change this number as needed

for i in range(min(n_trees_to_plot, len(rf_model.estimators_))):
    plt.figure(figsize=(16, 12))
    plot_tree(rf_model.estimators_[i], filled=True, feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], class_names=['setosa', 'versicolor', 'virginica'], rounded=True, fontsize=8)
    plt.title(f'Tree {i + 1} from Random Forest')
    plt.show()
```

```python
# Making predictions on the test set
y_hat = rf_model.predict(X_test)
y_hat
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_hat, y_test) # ROC AUC is for binary classifier
```
